# 코드

```python
# E:\down\업로드_temp\리뷰VOC분석* glob해서 df로 읽어오기
file_list = glob.glob(r"E:\down\업로드_temp\리뷰VOC분석*.xlsx")
df_list = []
for file in file_list:
    df_temp = pd.read_excel(file, header=6)
    df_list.append(df_temp)
df = pd.concat(df_list, ignore_index=True)
df.drop_duplicates(inplace=True)
df = df[df["감정수준"] == "부정"]
```

```python
df = df.copy()
df
```

```python
df = df[df["감정수준"] == "부정"]
```

```python
# 일자/매장별 부정수 집계
df = df.groupby(["작성일자", "매장명"], as_index=False).size().rename(columns={"size": "부정수"})

# rolling 전에 매장별 날짜 정렬 필수
df = df.sort_values(["매장명", "작성일자"]).reset_index(drop=True)

df
```

```python
# --- 30일 합계 계산 ---
# 최근 30일 합계
df['sum_30d_recent'] = df.groupby('매장명')['부정수'].transform(
    lambda x: x.rolling(window=30, min_periods=1).sum()
).astype(int)

# 이전 30일 합계 (31~60일 전)
df['sum_30d_prev'] = df.groupby('매장명')['부정수'].transform(
    lambda x: x.shift(30).rolling(window=30, min_periods=1).sum()
).fillna(0).astype(int)

df['pct_change_30d'] = (df['sum_30d_recent'] - df['sum_30d_prev']) / df['sum_30d_prev'].replace(0, np.nan) * 100
```

```python
# today변수선언 str로 
today = dt.datetime.now().strftime("%Y-%m-%d")
df = df[df["작성일자"] == today]
df = df[df["sum_30d_recent"] < df["sum_30d_prev"]]
```
