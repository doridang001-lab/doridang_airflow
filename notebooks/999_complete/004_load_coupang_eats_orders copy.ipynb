```python
import pandas as pd
from modules.transform.key_generator import add_surrogate_key
```

```python
def coupang_eates_sales_df():
    """
    ============================================
    REQ_ID: REQ001
    요청일: 2025-12-11
    요청자: 정민주 (E001)
    팀: 매출
    상태: 진행중
    ============================================
    """

    df = pd.read_excel(
        r"D:\Download\coupang_eats_2025-11.xlsx",
        header=1           # 2행을 헤더로 사용
    )
    df = df[1:]
    df = df[["주문번호", "일자", "시간", "주문금액"]].tail(10)
    df["일자"] = pd.to_datetime(df["일자"]).dt.date
    df = add_surrogate_key(df, natural_key_cols=["주문번호", "일자", "시간"])
    df["주문금액"] = df["주문금액"].astype(int)
    df.rename(columns={"key" : "order_id"}, inplace=True)
    return df
```
